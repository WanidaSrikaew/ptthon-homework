```python
class Dog():
    def __init__(self, name, gender, gene, age, weight, color):
        self.name = name
        self.gender = gender
        self.gene = gene

        # encapsulate
        self.__age = age #private attribute
        self.weight = weight
        self.color = color

    def setAge(self, newAge):
        if newAge < self.__age:
            print("Error !!")
        else:
            self.__age = newAge

    def showInfo(self):
        print("""
        ชื่อ: {}
        เพศ: {}
        สายพันธ์: {}
        อายุ: {}
        น้ำหนัก: {}
        สี: {}
        """.format(self.name, self.gender, self.gene, self.__age, self.weight, self.color))
    
    def sit(self):
        print("{} is sitting !!".format(self.name))

    def sleep(self):
        print("{} is sitting !!".format(self.name))

    def eat(self):
        print("{} is sitting !!".format(self.name))
```
