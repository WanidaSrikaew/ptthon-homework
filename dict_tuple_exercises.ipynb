## ข้อ 7: ค้นหาข้อมูลใน Dictionary

สร้าง dictionary ชื่อ `phone_book` ที่เก็บหมายเลขโทรศัพท์:
```python
phone_book = {
    "สมชาย": "081-234-5678",
    "สมหญิง": "089-876-5432",
    "สมศรี": "085-111-2222"
}
```

ให้สร้างฟังก์ชันที่ค้นหาหมายเลขโทรศัพท์ตามชื่อ และถ้าไม่พบให้แสดงข้อความ "ไม่พบข้อมูล"

````python
# เขียนคำตอบของคุณที่นี่
phone_book = {
    "สมชาย": "081-234-5678",
    "สมหญิง": "089-876-5432",
    "สมศรี": "085-111-2222"
}

def search_phone(name):
    return phone_book.get(name, "ไม่พบข้อมูล")
print(search_phone("สมชาย"))
print(search_phone("สมหญิง"))
print(search_phone("สมปอง"))
````

## ข้อ 9: การรวม Dictionary

สร้าง dictionary 2 อัน:
- `person1 = {"name": "สมชาย", "age": 25}`
- `person2 = {"name": "สมหญิง", "age": 24}`

ให้รวม dictionary ทั้งสองเป็น dictionary เดียวชื่อ `all_people` และพิมพ์ผลลัพธ์

หมายเหตุ: ลองใช้วิธีต่างๆ เช่น update(), {**dict1, **dict2} เป็นต้น

````python
# เขียนคำตอบของคุณที่นี่
person1 = {"name": "สมชาย", "age": 25}
person2 = {"name": "สมนึก", "age": 24}

all_people = {**person1, **person2}
print(all_people)
````

## ข้อ 10: โปรแกรมจัดการสินค้า (Challenge)

สร้างโปรแกรมจัดการสินค้าที่มีความสามารถดังนี้:

1. มี dictionary ชื่อ `inventory` ที่เก็บข้อมูลสินค้า: ชื่อสินค้า, จำนวน, และราคา (เก็บในรูป tuple)
2. สร้างฟังก์ชน `add_product()` เพื่อเพิ่มสินค้าใหม่
3. สร้างฟังก์ชั่น `update_quantity()` เพื่ออัปเดตจำนวนสินค้า
4. สร้างฟังก์ชั่น `get_total_value()` เพื่อคำนวณมูลค่ารวมของสินค้า
5. สร้างฟังก์ชั่น `show_inventory()` เพื่อแสดงรายการสินค้าทั้งหมด

ตัวอย่างการใช้งาน:
```
Inventory: {'น้ำตาล': (10, 50), 'แป้งมัน': (5, 100)}
Total value: 1000 บาท
```

````python
# เขียนคำตอบของคุณที่นี่
inventory = {
    'กล้วย': (10, 50, ['ผลไม้', 'สีเหลือง']),
    'นม': (5, 100, ['ผลิตภัณฑ์นม', 'เครื่องดื่ม'])
}

def add_product(name, quantity, price, tags=None):
    if name not in inventory:
        inventory[name] = (quantity, price, tags or [])
        print(f" เพิ่มสินค้า: {name} จำนวน {quantity} หน่วย")
    else:
        print(f" สินค้า '{name}' มีอยู่แล้ว")

def update_quantity(name, new_quantity):
    if name in inventory:
        price, tags = inventory[name][1], inventory[name][2]
        inventory[name] = (new_quantity, price, tags) 
        print(f" อัปเดตจำนวน '{name}' เป็น {new_quantity} หน่วย")
    else:
        print(f" ไม่พบสินค้า '{name}'")

def get_total_value(name):
    if name in inventory:
        quantity, price, _ = inventory[name]
        return quantity * price
    return 0

def show_inventory():
    print("\n---  รายการสินค้าในคลัง ---")
    grand_total = 0
    for name, (qty, price, _) in inventory.items():
        total_value = qty * price
        grand_total += total_value
        print(f"{name}: จำนวน {qty} หน่วย | ราคา/หน่วย {price} บาท | มูลค่ารวม {total_value} บาท")
    
    print(f"\n มูลค่ารวมทั้งหมดในคลัง: {grand_total} บาท")


show_inventory()

add_product('แอปเปิล', 20, 45, ['ผลไม้'])
update_quantity('กล้วย', 15)

print(f"มูลค่ารวมของนม: {get_total_value('นม')} บาท")

show_inventory()
````
